Import Library

```python
from graph.graph import Graph
from graph.edge import Edge
from graph.vertex import Vertex
from graph.graph_visualization import GraphVisualization
import networkx as nx
import pandas as pd
```

Model 1: Directed Graph

Step1 : Load Data

```python
df = pd.read_csv('model1.tsv', sep=' ')
print(df)
gr = Graph()
for d in df.iloc:
    src = str(d[0])
    des = str(d[1])
    gr.add_edge(Edge(Vertex(src), Vertex(des)))

```

Step2 : Visual Graph using Networkx

```python
gv = GraphVisualization(gr)
G1 = gv.networkxDiGraph()
#Customize the graph visualization
pos = nx.circular_layout(G1)
nx.draw(G1, pos, node_size=1000, node_color='black',font_color = 'white', edge_color='green', width= 2, with_labels=True)
# Display the graph 
```

Model 2: Undirected Graph

Step 1: Load data

```python
df = pd.read_csv('model2.tsv', sep=' ')
print(df)
gr1 = Graph()
for d in df.iloc:
    src = str(d[0])
    des = str(d[1])
    gr1.add_edge(Edge(Vertex(src), Vertex(des)))
```

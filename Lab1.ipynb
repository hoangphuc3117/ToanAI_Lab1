```python
from graph.graph import Graph
from graph.edge import Edge
from graph.vertex import Vertex
from graph.graph_visualization import GraphVisualization
import networkx as nx
import matplotlib.pyplot as plt
```

Model 1: Directed Graph

```python
gr = Graph()
edges = [] 
edges.append(Edge(Vertex('0'), Vertex('1')))
edges.append(Edge(Vertex('0'), Vertex('5')))
edges.append(Edge(Vertex('1'), Vertex('2')))
edges.append(Edge(Vertex('2'), Vertex('3')))
edges.append(Edge(Vertex('3'), Vertex('5')))
edges.append(Edge(Vertex('3'), Vertex('4')))
edges.append(Edge(Vertex('4'), Vertex('0')))
edges.append(Edge(Vertex('5'), Vertex('4')))
gr.add_edges(edges = edges)
es = gr.edges()
for e in es:
    print('($s $s)',e.v[0], e.v[1])

gv = GraphVisualization(gr)
G1 = gv.networkxDiGraph()
#Customize the graph visualization
pos = nx.circular_layout(G1)
plt.clf()
nx.draw(G1, pos, node_size=1000, node_color='black',font_color = 'white', edge_color='green', width= 2, with_labels=True)
# Display the graph 
plt.show()
```
